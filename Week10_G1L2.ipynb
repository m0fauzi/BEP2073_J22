```python
def f(m):
  g,Cd,t,v = 9.81,0.25,4,36
  return math.sqrt(g*m/Cd)*math.tanh(math.sqrt(g*Cd*t/m)) - v
```

```python
import scipy.optimize as so
x_sol = so.root_scalar(f, bracket = [1, 300], method = 'brentq')
```

```python
#(x-3)(x-2)=0, x1 = 3, x2 = 2
def f(x):
  return x**2 - 5*x + 6
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,20)
y = f(x)

plt.plot(x,y,'.r-')
plt.show()
```

```python
import scipy.optimize as so

x_opt = so.minimize_scalar(f)
print(x_opt.x)
```

```python
y_opt = f(x_opt.x)
plt.plot(x,y,'.r-')
plt.plot(x_opt.x,y_opt,'sk')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_cf.csv')

X = data.x
Y = data.y

plt.plot(X,Y,'or')
plt.show()
```

```python
import scipy.optimize as so
import numpy as np

def f(x,a1,a2,a3):
  return a1*x + a2/x + a3*np.log(x)

a_sol, a_cov = so.curve_fit(f, X, Y)

#a1 = a_sol[0]
#a2 = a_sol[1]
#a3 = a_sol[2]

print(a_sol)
```

```python
a1,a2,a3 = a_sol

x = np.linspace(0,6,100)
y = f(x,a1,a2,a3)

plt.plot(X,Y,'or')
plt.plot(x,y,'-k')
plt.show()
```

```python
# Find x = 1.5?

ans = f(1.5,a1,a2,a3)
print(ans)
```

```python
plt.plot(X,Y,'or')
plt.plot(x,y,'-k')
plt.plot(1.5,ans,'sb')
plt.show()
```
