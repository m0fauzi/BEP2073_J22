```python
import math
import pandas as pd
data = pd.read_csv('data_random.csv')
x = data.X
y = data.Y

def f(x,y):
  return 2*math.sin(x) + 3*math.cos(y)

for idx in [3, 8, 10, 15]: 
  z = f(x[idx],y[idx])
  print('At index %d: z = %0.3f' %(idx, z))
```

```python
#Plot x vs y

import matplotlib.pyplot as plt
plt.plot(x,y,'.r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import scipy.optimize as so
def f(x):
  return x**2 - 2*x - 35
ans = so.root_scalar(f, bracket = [-8,0], 
                     method = 'brentq')
print(ans.root)  
```

image.png

```python
ans = so.minimize_scalar(f)
x_opt = ans.x
f_opt = ans.fun
```

```python
import numpy as np

x = np.linspace(-10,10)
y = f(x)

plt.plot(x,y,'-r')
plt.plot(x_opt,f_opt,'sk')
plt.show()
```

```python
data = pd.read_csv('data_cf.csv')

x = data.x
y = data.y
plt.plot(x,y,'.r')
plt.show()

def f(x,a1,a2,a3):
  return a1*x + a2/x + a3*np.log(x)

a,cov = so.curve_fit(f,x,y)
print(a[0])
print(a[1])
print(a[2])
```

```python
X = np.linspace(0,6,30)
a1,a2,a3 = a
Y = f(X,a1,a2,a3)

x = data.x
y = data.y
plt.plot(x,y,'.r')
plt.plot(X,Y,'-k')
plt.show()
```
