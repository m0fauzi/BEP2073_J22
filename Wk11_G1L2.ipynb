```python
def Cp(T,a1,a2,a3,a4,a5):
  a0 = 85.6
  return a0 + a1*T + a3*T**2 + a4*T**3 + a5*T**4
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('data6_2.csv')

x = data.t #ind
y = data.c #dep

plt.plot(x,y,'r*')
plt.plot([0,25],[5, 5],'--k')
plt.xlabel('t')
plt.ylabel('c')
plt.show()
```

```python
def Conc(t,a0,a1,a2,a3,a4):
  return a0 + a1*t + a3*t**2 + a4*t**3 + a5*t**4 
```

```python
import scipy.optimize as so

a,cov = so.curve_fit(Conc,x,y)
```

```python
import numpy as np

tdata = np.linspace(0,25,100)
a0,a1,a2,a3,a4 = a
ydata = Conc(tdata,a0,a1,a2,a3,a4)
```

```python
plt.plot(x,y,'r*')
plt.plot([0,25],[5, 5],'--k')
plt.plot(tdata,ydata,'-b')
plt.xlabel('t')
plt.ylabel('c')
plt.show()
```
